```python
import pandas as pd
import wget
#urls='https://github.com/CSSEGISandData/COVID-19/raw/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv'        
#wget.download(urls)
```

```python
confirmed_df=pd.read_csv('time_series_covid19_confirmed_global.csv')
deaths_df=pd.read_csv('time_series_covid19_deaths_global.csv')
recovered_df=pd.read_csv('time_series_covid19_recovered_global.csv')
```

```python
#all the dates columns are putting into a variable called dates
date=confirmed_df.columns[4:]
confirmed_df_long=confirmed_df.melt(
    id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'],
    value_vars=date,
    var_name='Date',
    value_name='Confirmed'
)

deaths_df_long=deaths_df.melt(
    id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'],
    value_vars=date,
    var_name='Date',
    value_name='Deaths'
)

recovered_df_long=recovered_df.melt(
    id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'],
    value_vars=date,
    var_name='Date',
    value_name='Recovered'
)
```

```python
confirmed_df_long.head(-10)
```

```python
#merge three different dataset
#merge confirmed_df_long and death_df_long
full_table= confirmed_df_long.merge(
    right= deaths_df_long,
    how= 'left',
    on=['Province/State','Country/Region','Lat','Long','Date']
)

#merge full_table with recovered_df_long
full_table=full_table.merge(
    right=recovered_df_long,
    how='left',
    on=['Province/State','Country/Region','Lat','Long','Date']
)

full_table.head(100)
```

```python
#data cleaning
#convertig date from string to date format
full_table['Date']=pd.to_datetime(full_table['Date'])
full_table.head()
```

```python
#finding missing values
full_table.isna().sum()
```

```python
full_table['Recovered']=full_table['Recovered'].fillna(0)
```

```python
full_table.isna().sum()
```

```python
#no of active cases=confirmed-death-recovered
full_table['Active']=full_table['Confirmed']-full_table['Deaths']-full_table['Recovered']
full_table.head(75)
```

```python
full_grouped = full_table.groupby(['Date','Country/Region'])[['Confirmed','Deaths','Recovered','Active']].sum().reset_index()
```

```python
full_grouped.head(50)
```

```python
import numpy as np
full_grouped = full_table.groupby(['Date','Country/Region'])[['Confirmed','Deaths','Recovered','Active']].sum().reset_index()
temp=full_grouped.groupby(['Country/Region','Date'])[['Confirmed','Deaths','Recovered']]
temp=temp.sum().diff().reset_index()

mask=temp['Country/Region'] != temp['Country/Region'].shift(1)
temp.loc[mask,'Confirmed']=np.nan
temp.loc[mask,'Deaths']=np.nan
temp.loc[mask,'Recovered']=np.nan
#rename column
temp.columns=['Country/Region','Date','New cases','New Deaths','New recovered']

full_grouped=pd.merge(full_grouped,temp,how='left',on=['Country/Region','Date'])
#fill na with 0
full_grouped=full_grouped.fillna(0)
#fixing datatypes
cols=['New cases','New Deaths','New recovered']
full_grouped[cols]=full_grouped[cols].astype(int)
full_grouped['New cases']=full_grouped['New cases'].apply(lambda x:0 if x<0 else x)
```
